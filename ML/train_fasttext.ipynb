```python

import fasttext

training_data_path = '/Users/linh/project/kltn/ml/data/cooking.train'

model = fasttext.train_unsupervised(
    input=training_data_path,
    model='cbow',       # 'cbow' or 'skipgram'
    dim=300,            # Dimension of word vectors
)

model.save_model('/Users/linh/project/kltn/ml/data/trained_model_dim300.bin')
```
